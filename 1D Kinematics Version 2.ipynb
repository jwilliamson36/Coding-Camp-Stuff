```
# These are some software packages we'll use to allow us to analyze this problem (Remember to run after each code section is complete)
import pandas as pd
import numpy as np
from math import cos,sin,pi
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# POSITION vs TIME GRAPHS

```
#Defining Lists and Equations used for Superman
xi = 250000    #the initial position of Superman
xf = 0   #the final position of Superman
vs = -455      #the velocity of Superman
sa = 0        #the acceleration of Superman
tdelta = 10   #the periods of time we will calculate data
t = 0         #the initial time
S = xi    #the position of Superman at any moment in time
Sv = vs   #the velocity of Superman at any moment in time
tlist = [] #create a list for the time values
Slist = [] #create a list for the position values
Svlist = [] #create a list for the velocity values
tlist.append(t)
Slist.append(S)
Svlist.append(Sv)



while S >= 0: #stop the calculations after Superman reaches Exit 0
    S = .5*sa*tdelta*tdelta + vs*tdelta + S #calculating the position at any moment in time
    Slist.append(S)

    Sv = sa*tdelta + Sv #calculating the velocity at any moment in time
    Svlist.append(Sv)

    t = t + tdelta #calculating the time at each interval
    tlist.append(t)

Sdata = pd.DataFrame({"t": tlist, "S": Slist, "Sv": Svlist})
Sdata.head(5) # only see first five rows of data
```

```
#What does the data for Superman look like?  Create a position time graph for Superman

plt.plot(Sdata['t'],Sdata['S'], linestyle='--', marker='o', color='b', markersize=2)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Superman Position vs Time')
plt.grid(True)
plt.show()

```

```
#Defining Lists and Equations used for Flash
Fxi = 32000 #originial position of Flash
Fxf = 202000  #final position of Flash
Fvi = 0   #original velocity of Flash
Fa = 3.78   #acceleration of Flash
tdelta = 30  #length of time intervals
t = 0   #original time
Fv = Fvi   #the velocity of the Flash
F = Fxi    #the position of the Flash
tlist = []
Flist = []
Fvlist = []
tlist.append(t)
Flist.append(F)
Fvlist.append(Fv)


while F <= 202000:

    F = .5*Fa*tdelta*tdelta + Fv*tdelta + F
    Flist.append(F)

    Fv = Fa*tdelta + Fv
    Fvlist.append (Fv)
    t = t + tdelta
    tlist.append(t)

Fdata = pd.DataFrame({"t": tlist, "F": Flist, "Fv": Fvlist})
Fdata.head (5)
```

```
#What does the graphical representation of the Flash look like?
plt.plot(Fdata['t'],Fdata['F'], linestyle='--', marker='o', color='r', markersize=2)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('The Flash Position vs Time')
plt.grid(True)
plt.show()
```

```
fig, ax = plt.subplots(2, 2, figsize=(10,5))
fig.subplots_adjust(hspace=0.5)

ax[0,0].plot(Sdata['t'], Sdata['Sv'], color= 'blue', linestyle="--")
ax[0,0].set(xlabel="Time (s)", ylabel="Velocity (m/s)", title="Superman Velocity")

ax[0,1].plot(Fdata['t'], Fdata['Fv'], color= 'red', linestyle="--")
ax[0,1].set(xlabel="Time (s)", ylabel="Velocity (m/s)", title="Flash Velocity")


ax[1,0].plot(Sdata['t'], Sdata['Sv'], color= 'blue', label='Superman', linestyle="--")
ax[1,0].plot(Fdata['t'], Fdata['Fv'], color= 'red', label='Flash', linestyle="--")
ax[1,0].set(xlabel="Time (s)", ylabel="Velocity (m/s)", title="Superman and Flash Velocity")
ax[1,0].legend()

ax[1,1].plot(Sdata['t'], np.abs(Sdata['Sv']), color= 'blue', label='Superman', linestyle="--")
ax[1,1].plot(Fdata['t'], Fdata['Fv'], color= 'red', label='Flash', linestyle="--")
ax[1,1].set(xlabel="Time (s)", ylabel="Speed (m/s)", title="Superman and Flash Speed")
ax[1,1].legend();
```
